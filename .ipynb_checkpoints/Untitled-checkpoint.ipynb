```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': [1, 1, 2, 1, 2],
                    'B': [np.nan, 2, 3, 4, 5],
                    'C': [1, 2, 1, 1, 2]}, columns=['A', 'B', 'C'])
df
```

```python
# Groupby one column and return the mean of the remaining columns in each group.
# The values in column A are sorted into 1 or 2
# For all 1 values in A, the mean of corresponding B values is 3.0
# For all 1 values in A, the mean of corresponding C values is 1.3333
# For all 2 values in A, the mean of corresponding B values is 4.0
# For all 2 values in A, the mean of corresponding C values is 1.5
df = df.groupby('A').mean()
df
```
